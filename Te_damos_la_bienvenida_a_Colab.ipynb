```
def dividir_seguro(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: division entre cero"

print(dividir_seguro(10, 2))
print(dividir_seguro(8, 0))
print(dividir_seguro(5, 2))
```

```
colores = ["rojo", "azul", "rojo", "verde", "azul", "rojo"]
conteo = {}

for color in colores:
    if color in conteo:
        conteo[color] = conteo[color] + 1
    else:
        conteo[color] = 1

print(f"rojo: {conteo["rojo"]}")
print(f"azul: {conteo["azul"]}")
print(f"verde: {conteo["verde"]}")
```

```
notas = {
    "Pedro": 78,
    "Sofia": 95,
    "Luis": 82,
    "Maria": 88
}

mejor_nombre = ""
mejor_nota = 0

for nombre in notas:
    if notas[nombre] > mejor_nota:
        mejor_nota = notas[nombre]
        mejor_nombre = nombre

print(f"Mejor alumno: {mejor_nombre} con {mejor_nota}")
```

```
def invertir(texto):
    return texto[::-1]

print(invertir("Python"))
print(invertir("racecar"))

palabra = "racecar"
es_palindromo = palabra == invertir(palabra)
print(f"racecar es palindromo: {es_palindromo}")
```
